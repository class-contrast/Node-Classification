```python
Node_Year=D['node_year']['paper']
Node_Year_list=list(Node_Year[:,0])

E=D['edge_index_dict'][('paper', 'cites', 'paper')]
EdgeL=E[0]
EdgeR=E[1]
print(EdgeL)
print(EdgeR)
print(np.unique(Node_Year_list))

Node_Year_list.index(2019)
list_size = len(Node_Year_list)
NIndex_19=[]
# declare for loop
for itr in range(list_size):
 
      # check the condition
    if(Node_Year_list[itr] == 2019):
        NIndex_19.append(itr)
 
          # print the indices
print(len(NIndex_19))
```

```python
N_class_train=Node_Class
for d in NIndex_19:
    N_class_train[d]=349
print(np.unique(N_class_train))
```

```python
def CountX(lst,x):
    return(lst.count(x))

Node_class=list(range(350))
F_vec=[]
n=736389
for i in range(n):
    print("\r Processing file {} ({}%)".format(i, 100 * i //n), end='', flush=True)
    node_F=[]
    list_out=[]
    list_In=[]
    #S_nbd_out=[]
    #S_nbd_in=[]
    indx=np.where(EdgeL==i)
    indxIn=np.where(EdgeR==i)
    #print(indx)
    for j in indx[0]:
        list_out.append(N_class_train[EdgeR[j]])
   # indx[0].clear()
    for k in indxIn[0]:
        list_In.append(N_class_train[EdgeL[k]])
   # indxIn[0].clear()

                
                
    for d in Node_class:
        node_F.append(CountX(list_out,d))
        node_F.append(CountX(list_In,d))
    F_vec.append(node_F)
```

```python
n=736388

node_F=[]
list_out=[]
list_In=[]
    #S_nbd_out=[]
    #S_nbd_in=[]
indx=np.where(EdgeL==n+1)
indxIn=np.where(EdgeR==n+1)
    #print(indx)
for j in indx[0]:
    list_out.append(N_class_train[EdgeR[j]])
   # indx[0].clear()
for k in indxIn[0]:
    list_In.append(N_class_train[EdgeL[k]])
   # indxIn[0].clear()

                
for d in Node_class:
    node_F.append(CountX(list_out,d))
    node_F.append(CountX(list_In,d))
F_vec.append(node_F)
```

```python
del F_vec[-1]
print(len(F_vec))
print(len(Node_Class))
```

```python
feature=['0']
for i in range(1,700):
    feature.append("{}".format(i))
data = pd.DataFrame(F_vec, columns =feature) 
data.insert(loc=700,column='Class',value=Node_Class)
data.head()
```

```python
compression_opts = dict(method='zip',
                        archive_name='Feature_1it_nbd_mag.csv')  
data.to_csv('Feature_1it_nbd_mag.zip', index=True,
          compression=compression_opts) 
```

```python
result=pd.concat([data, df2],axis=1)
result.head()
```

```python
feature=['0']
for i in range(1,700):
    feature.append("{}".format(i))
for i in range(128):
    feature.append("w_{}".format(i))

X=result[feature] # Features
y=result['Class']  # Labels
X_train=X.iloc[list(train_idx['paper'])+list(valid_idx['paper'])]
X_test=X.iloc[test_idx['paper']]
y_train=y.iloc[list(train_idx['paper'])+list(valid_idx['paper'])]
y_test=y.iloc[test_idx['paper']]
from xgboost import XGBClassifier

bst = XGBClassifier(n_estimators=100, max_depth=3, learning_rate=0.1)
bst.fit(X_train,y_train)
y_pred=bst.predict(X_test)
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred)*100,"%")
```
