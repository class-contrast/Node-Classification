```python
import pandas as pd
import numpy as np
import ogb
from ogb.nodeproppred import NodePropPredDataset 
from torch_geometric.data import DataLoader
#H:\\PhD Research\Data\PROTEINS\PROTEINS_node_attributes.txt
# Download and process data at './dataset/ogbg_molhiv/'
dataset = NodePropPredDataset(name = "ogbn-proteins", root = '/Users/joshem/PhD Research/Data/')
split_idx = dataset.get_idx_split()
train_idx, valid_idx, test_idx = split_idx["train"], split_idx["valid"], split_idx["test"]
graph = dataset[0] # pyg graph object
```

```python
import pandas as pd
import numpy as np
D=graph[0]
E=graph[1]
```

```python
import time
def CountX(lst,x):
    return(lst.count(x))
Node_class=np.unique(D['node_species'])
#print(Node_class)
Edge_ind=D['edge_index']
Edge_feat=D['edge_feat']
Node_species=D["node_species"]
F_vec=[]
F_vec1=[]
n=132534
#n=100
start_time = time.time()
for i in range(1,n+1):
  print("\rProcessing file {} ({}%)".format(i, 100*i//n), end='', flush=True)
  node_F=[]
  #List_class=[]
  V1120=[]
  V112=[]
  V1121=[]
  V1122=[]
  V1123=[]
  V1124=[]
  V1125=[]
  V1126=[]
  V1127=[]
  indx=np.where(Edge_ind[0]==i)
  #print(indx[0])
  for j in indx[0]:
    #List_class.append(Node_species[Edge_ind[1][j]][0])
    V1120.append(E[Edge_ind[1][j]])
    V112.append(Edge_feat[j][0]*E[Edge_ind[1][j]])
    V1121.append(Edge_feat[j][1]*E[Edge_ind[1][j]])
    V1122.append(Edge_feat[j][2]*E[Edge_ind[1][j]])
    V1123.append(Edge_feat[j][3]*E[Edge_ind[1][j]])
    V1124.append(Edge_feat[j][4]*E[Edge_ind[1][j]])
    V1125.append(Edge_feat[j][5]*E[Edge_ind[1][j]])
    V1126.append(Edge_feat[j][6]*E[Edge_ind[1][j]])
    V1127.append(Edge_feat[j][7]*E[Edge_ind[1][j]])
  #print(List_class)
  #print(C)
  #for d in Node_class:
  #     node_F.append(CountX(List_class,d))
  F_vec1.append([sum(x) for x in zip(*V1120)]+[sum(x) for x in zip(*V112)]+[sum(x) for x in zip(*V1121)]+[sum(x) for x in zip(*V1122)]+[sum(x) for x in zip(*V1123)]+[sum(x) for x in zip(*V1124)]+[sum(x) for x in zip(*V1125)]+[sum(x) for x in zip(*V1126)]+[sum(x) for x in zip(*V1127)])
  #F_vec.append(node_F)
end_time = time.time()
print((end_time-start_time)/3600, 'hours')
```
