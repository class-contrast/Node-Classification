```python
#READ THE DATA AND CREATE ADJACENCY MATRIX

import numpy as np
import pandas as pd
p_data = open("/Users/joshem/PhD Research/Data/citeseer-doc/citeseer.cites")


# Create a list from the data
myls=list(p_data)

# Remove the new line symbole from the list
mylist = [line.rstrip('\n') for line in myls]
#print(mylist)
#print(graph_ind)
#print(graph_level)

p_data.close()
mynode=[]
i=0
for d in mylist:
    d=mylist[i].split("\t")
    mynode.append(d[0])
    mynode.append(d[1])
    i=i+1
#print(mynode)
node_ID1=np.unique(mynode)
node_ID2=list(node_ID1)
#print(node_ID2)
n=len(node_ID2)
print(n)
from natsort import natsorted, ns
node_ID=natsorted(node_ID2)
#print(node_ID)
A=np.zeros((n,n))
i=0
for d in mylist:
    d=mylist[i].split("\t")
    A[node_ID.index(d[0])][node_ID.index(d[1])]=1
    i=i+1;
print(" The adjacency Matrix")
print("=======================")
print(A)
```

```python
#READ THE NODE FEATURE

import os
data_dir = os.path.expanduser("/Users/joshem/PhD Research/Data/citeseer-doc")

feature_names = ["w_{}".format(ii) for ii in range(3703)]
column_names = ["ID"]+ feature_names + ["subject"]
node_data = pd.read_csv(os.path.join(data_dir, "citeseer.content"), sep='\t', header=None, names=column_names)
node_data.head()
```

```python
DataS=node_data.replace({'subject': {'Agents':0 , 'AI': 1,'DB':2,'IR':3 , 'ML': 4,'HCI':5}})
# Agents==0
#		AI==1
#		DB==2
#		IR==3
#		ML==4
#		HCI==5

DataS.head()
```

```python
#SORT THE NODE FEATURE ACCORDING TO ID

from natsort import index_natsorted, order_by_index

Data = DataS.reindex(index=order_by_index(DataS.index, index_natsorted(Data['ID'], reverse=False)))
#Data = Data.reset_index(drop=True)
Data.head()
```

```python
#RANDOMLY CHOOSE 1000 NODE FOR TEST DATA

catagories=Data['subject'].to_numpy()
print(catagories)
import random
p=random.sample(range(3312), 1000)
#print(p)
q=[]
for i in range(n):
    if i not in p:
        q.append(i)
```

```python
#LABLE THE UNKNOWN/TEST NODE AS 6 AND PLACE CLASS TO THE DIAGONAL OF ADJACENCY MATRIX

class_60=np.array(catagories)
for d in p:
    class_60[d]=6
print(class_60)
for i in range(n):
    A[i][i]=class_60[i]
print(A)
```

```python
def CountX(lst,x):
    return(lst.count(x))
#CREARTE AN ARRAY OF 1 AND 2 HOP NBD FOR EACH NODE

Node_class=[0,1,2,3,4,5]
F_vec=[]
for i in range(n):
    node_F=[]
    list_out=[]
    list_In=[]
    S_nbd_out=[]
    S_nbd_in=[]
    for j in range(n):
        if A[i][j]==1 and i!=j:
            list_out.append(A[j][j])
            for k in range(n):
                if A[j][k]==1 and k!=j:
                    S_nbd_out.append(A[k][k])
        if A[j][i]==1 and i!=j:
            list_In.append(A[j][j])
            for k in range(n):
                if A[k][j]==1 and k!=j:
                    S_nbd_in.append(A[k][k])
     # COUNT FOR EACH CLASS           
    for d in Node_class:
        node_F.append(CountX(list_out,d))
        node_F.append(CountX(list_In,d))
    for d in Node_class:
        node_F.append(CountX(S_nbd_out,d))
        node_F.append(CountX(S_nbd_in,d))
    F_vec.append(node_F)
```

```python
x =np.array(F_vec)
k=24
feature=["a_0","b_0"]
for i in range(1,k-1,2):
    feature.append("a_{}".format(int((i+1)/2)))
    feature.append("b_{}".format(int((i+1)/2)))

    
data = pd.DataFrame(x,columns=feature)
data.insert(loc=k,column='Class',value=Data['subject'].to_numpy())
data.head()
```
